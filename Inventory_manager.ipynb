```
import numpy as np
```

```
np.set_printoptions(linewidth=160)
garage=np.zeros([12,36],dtype=int)
garage
```

```
# initialize all the garage as path
garage=np.where(garage==0,-1,garage)
garage
```

```
# -1 are path, 0 are empty racks, 1 are full racks
garage[2:10,0]=0
garage[2:10,35]=0
for i in range(3,33,4):
  garage[2:10,i:i+2]=0
garage
```

```
# Now we fill up some random racks to get a garage structure with randomly distributed empty racks
import random
from random import randint
empty_racks=np.where(garage==0)
for i in range(0,len(empty_racks[0])-20): # here 20 is the number of racks which will be empty
  temp=random.randint(0,len(empty_racks[0])-1) # select a random rack from empty racks
  garage[empty_racks[0][temp],empty_racks[1][temp]]=1 # fill up the random rack
  empty_racks=np.where(garage==0) # update empty racks
garage
```

```
# Confirming how many empty racks left
len(empty_racks[0])
```

```
import matplotlib.pyplot as plt
garage[0][0]=-2
plt.matshow(garage,extent=[0,36,12,0])
plt.xticks(np.arange(0,37,1))
plt.yticks(np.arange(0,13,1))
plt.grid(color='black')
plt.savefig("warehouse.jpg")

# blue is the path, green are the empty racks, yellow are the full racks, and purple is the starting and ending position.
```

```
import networkx as nx
G=nx.Graph()
weights=[]
for i in range(20):
  if(empty_racks[1][i]+1==len(garage[0])):
    wt_x=empty_racks[1][i]-1
  elif(garage[empty_racks[0][i],empty_racks[1][i]+1]==-1):
    wt_x=empty_racks[1][i]+1
  else:
    wt_x=empty_racks[1][i]-1
  wt_y=empty_racks[0][i]
  G.add_edge('(0,0)' , '('+str(empty_racks[0][i])+','+str(empty_racks[1][i])+')' , weight=wt_x+wt_y)
```

```
plt.figure(1,figsize=(8,8))
color_map=['blue']*(len(empty_racks[0])+1)
color_map[0]='red'
nx.draw(G,with_labels=True,font_size=15,font_weight='bold',node_size=3000,node_color=color_map)
plt.show()
```

```
for i in range(len(empty_racks[0])):
  for j in range(i+1,len(empty_racks[0])):
    if(empty_racks[1][j]==empty_racks[1][i]):
      wt_x=0
      wt_y=abs(empty_racks[0][j]-empty_racks[0][i])

    elif(abs(empty_racks[1][j]-empty_racks[1][i])==3):
      wt_x=1
      wt_y=abs(empty_racks[1][j]-empty_racks[1][i])

    elif(empty_racks[1][i]==len(garage[0])-1):
      continue

    elif(empty_racks[1][j]==len(garage[0])-1 and garage[empty_racks[0][i],empty_racks[1][i]+1]==-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i]-2)
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    elif(empty_racks[1][j]==len(garage[0])-1 and garage[empty_racks[0][i],empty_racks[1][i]+1]!=-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i])
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    elif(garage[empty_racks[0][i],empty_racks[1][i]+1]==-1 and garage[empty_racks[0][j],empty_racks[1][j]+1]==-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i])
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    elif(garage[empty_racks[0][i],empty_racks[1][i]+1]==-1 and garage[empty_racks[0][j],empty_racks[1][j]+1]!=-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i]-2)
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    elif(garage[empty_racks[0][i],empty_racks[1][i]+1]!=-1 and garage[empty_racks[0][j],empty_racks[1][j]+1]!=-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i])
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    elif(garage[empty_racks[0][i],empty_racks[1][i]+1]!=-1 and garage[empty_racks[0][j],empty_racks[1][j]+1]==-1):
      wt_x=abs(empty_racks[1][j]-empty_racks[1][i]+2)
      wt_y=min(empty_racks[0][i]-1+empty_racks[0][j]-1 , len(garage)-empty_racks[0][i]-2+len(garage)-empty_racks[0][j]-2)

    G.add_edge('('+str(empty_racks[0][i])+','+str(empty_racks[1][i])+')' , '('+str(empty_racks[0][j])+','+str(empty_racks[1][j])+')' , weight=wt_x+wt_y)
```

```
plt.figure(1,figsize=(10,10))
nx.draw(G,with_labels=True,font_size=15,font_weight='bold',node_size=3000,node_color=color_map)
plt.show()
plt.savefig("all_empty_racks.png")
```

```
nodes=['(0,0)']
for i in range(len(empty_racks[0])):
  newnode='('+str(empty_racks[0][i])+','+str(empty_racks[1][i])+')'
  nodes.append(newnode)

nodes
```

```
from pylab import *

sub=G.subgraph(nodes=nodes[0:6])
newpath = nx.minimum_spanning_tree(sub, algorithm='prim')
color_map=['blue']*6
color_map[0]='red'
print(sub.nodes())

plt.figure(1,figsize=(25,12))
subplot(1,2,1)
nx.draw(sub,with_labels=True,font_size=15,font_weight='bold',node_size=3000,node_color=color_map)
subplot(1,2,2)
nx.draw(newpath,with_labels=True,font_size=15,font_weight='bold',node_size=3000,node_color=color_map)
plt.show()
```

```
# The generalized algorithm

from itertools import combinations
import math
def sett(arr, r): 
    return list(combinations(arr, r)) 

boxes=50
rack_size=10
racks_to_visit=math.ceil(boxes/rack_size)
color_map=['blue']*(racks_to_visit+1)
color_map[1]='red'

possible_visits=sett(nodes[1:21], racks_to_visit)
```

```
dist=[]
for i in range(len(possible_visits)):
  temp_nodes=possible_visits[i]
  temp_nodes = ['(0,0)']+list(temp_nodes)
  subgrph = G.subgraph(nodes=temp_nodes)
  path = nx.minimum_spanning_tree(subgrph)
  temp_dist = 0
  for (u,v,c) in path.edges.data('weight'):
    temp_dist += c
  for (u1,v1,c1) in G.edges.data('weight'):
    if(u1=='(0,0)' and v1==v):
      temp_dist += c1
  dist.append(temp_dist)
```

```
min_index=dist.index(min(dist))
print('Minimum distance is = ' , min(dist) , ' for the subgraph = ' , min_index) 
```

```
from pylab import *

temp_nodes=['(0,0)']+list(possible_visits[min_index])
subgrph = G.subgraph(nodes=temp_nodes)
path = nx.minimum_spanning_tree(subgrph)
for (u,v,c) in path.edges.data('weight'):
  print('Node1 = ', u, ' Node2 = ', v, ' Weight = ', c)
```
